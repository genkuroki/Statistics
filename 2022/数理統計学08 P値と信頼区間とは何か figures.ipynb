```julia
using Distributions
using StatsPlots
default(fmt=:png)

n,  = 80, 0.3
L = 10^5
Z = zeros(L)
for i in 1:L
    K = rand(Binomial(n, ))
    Z[i] = (K/n - ) / (*(1 - )/n)
end

Kbin = -0.5:n+0.5
bin = @. (Kbin/n - ) / (*(1 - )/n)
histogram(Z; norm=true, alpha=0.3, bin, label="(K/n - ) / (*(1 - )/n)")
plot!(Normal(0, 1); label="Normal(0, 1)", lw=2)
plot!(xguide="z", legendfontsize=12)
title!("n=$n,  =$,  K ~ Binomial(n, )")
```

```julia
using Distributions
using StatsPlots
default(fmt=:png)
r(x) = round(x; sigdigits=3)
safediv(x, y) = x == 0 ? zero(x/y) : x/y

function pvalue_wilson(k, n, )
    a = safediv(k/n - , (*(1 - )/n))
    2ccdf(Normal(0, 1), abs(a))
end

n, k = 100, 60
plot( -> pvalue_wilson(k, n, ), 0, 1; label="pvalue_wilson(k, n, )")
plot!(xguide="", yguide="P-value")
plot!(legend=:topleft, legendfontsize=12)
plot!(xtick=0:0.1:1, ytick=0:0.05:1)
title!("Wilson's P-value function for n=$n, k=$k")
```
