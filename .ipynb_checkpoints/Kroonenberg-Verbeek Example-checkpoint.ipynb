# Kroonenberg-Verbeek Example

* P. M. Kroonenberg and Albert Verbeek, The Tale of Cochran's Rule: My Contingency Table has so Many Expected Values Smaller than 5, What Am I to Do?, The American Statistician, Volume 72, 2018 - Issue 2, 175-183. [web](https://www.tandfonline.com/doi/full/10.1080/00031305.2017.1286260)

```julia
using Distributions
using SymPy: sympy, @vars, oo
using Plots

# legendの半透明化
default(:bglegend, plot_color(default(:bg), 0.5))
default(:fglegend, plot_color(ifelse(isdark(plot_color(default(:bg))), :white, :black), 0.6));

using Base64
displayfile(mime, file; tag="img") = open(file) do f
    base64 = base64encode(f)
    display("text/html", """<$(tag) src="data:$(mime);base64,$(base64)"/>""")
end

pyplotclf() = if backend() == Plots.PyPlotBackend(); PyPlot.clf(); end 

pyplot(fmt=:svg)
```

```julia
safediv(x, y)  = iszero(x) ? x : x/y
safemult(x, y) = iszero(x) ? x : x*y

function chisq(A)
    r, c = size(A)
    n = sum(A)
    M = vec(sum(A, dims=2))
    N = vec(sum(A, dims=1))
    sum(safediv((A[i,j] - M[i]*N[j]/n)^2, M[i]*N[j]/n) for i in 1:r, j in 1:c)
end

function gstat(A)
    r, c = size(A)
    n = sum(A)
    M = vec(sum(A, dims=2))
    N = vec(sum(A, dims=1))
    2sum(safemult(A[i,j], log(A[i,j]) - log(M[i]*N[j]/n)) for i in 1:r, j in 1:c)
end
        
param_indep(p, q) = p .* q'
df_chisq(r, c) = (r-1)*(c-1)
df_chisq(P) = prod(size(P) .- 1)

chisq_pval(A) = ccdf(Chisq(df_chisq(A)), chisq(A))
gstat_pval(A) = ccdf(Chisq(df_chisq(A)), gstat(A))

function fisher_pval(hg::Hypergeometric, a::Integer; δ=1.0)
    s, f, n = params(hg)
    kmin = max(0, n - f)
    kmax = min(n, s)
    p0 = pdf(hg, a)
    p1 = 0.0
    pval = 0.0
    for k in kmin:kmax
        p1 = pdf(hg, k)
        pval += ifelse(p1 == p0, δ*p1, ifelse(p1 < p0, p1, 0.0))
    end
    return min(pval, 1)
end

function fisher_pval(A::AbstractArray{T,2}; δ=1.0) where T<:Integer
    a, b, c, d = vec(A)
    fisher_pval(Hypergeometric(a+b, c+d, a+c), a; δ=δ)
end

fisher_midp(A) = fisher_pval(A; δ=0.5)
```

```julia
function rand_Poissons(Λ)
    dist = product_distribution(Poisson.(vec(Λ)))
    reshape(rand(dist), size(Λ))
end

function rand_multinomial(Λ)
    n = round.(Int, sum(Λ))
    dist = Multinomial(n, vec(Λ)/n)
    reshape(rand(dist), size(Λ))
end

function rand_prod_multinomials(Λ)
    r, c = size(Λ)
    M = round.(Int, vec(sum(Λ, dims=2)))
    Q = Λ ./ M
    A = Array{Int, 2}(undef, r, c)
    for i in 1:r
        A[i,:] = rand(Multinomial(M[i], Q[i,:]))
    end
    A
end

rand_2binomials(Λ) = rand_prod_multinomials(Λ)

function rand_hypergeometric(Λ)
    a, c, b, d = vec(Λ)
    s, f, n = round.(Int, (a+c, b+d, a+b))
    k = rand(Hypergeometric(s, f, n))
    [
        k   n-k
        s-k k+f-n
    ]
end

function rand_noncentral_hypergeom(Λ)
    a, c, b, d = vec(Λ)
    s, f, n, ω = a+c, b+d, a+b, (a*d)/(b*c)
    k = rand(FisherNoncentralHypergeometric(s, f, n, ω))
    [
        k   n-k
        s-k k+f-n
    ]
end

function sim2x2(rand2x2; L=10^5)
    Chisq_Pval  = Array{Float64, 1}(undef, L)
    Gstat_Pval  = Array{Float64, 1}(undef, L)
    Fisher_Pval = Array{Float64, 1}(undef, L)
    Fisher_midP = Array{Float64, 1}(undef, L)
    for l in 1:L
        A = rand2x2()
        Chisq_Pval[l]  = chisq_pval(A)
        Gstat_Pval[l]  = gstat_pval(A)
        Fisher_Pval[l] = fisher_pval(A)
        Fisher_midP[l] = fisher_midp(A)
    end
    Chisq_Pval, Gstat_Pval, Fisher_Pval, Fisher_midP
end

sim2x2(rand_func, Λ; L=10^5) = sim2x2(() -> rand_func(Λ); L=L)
```

```julia
ecdf(x; Y=randn(100)) = mean(Y .≤ x)

function plot_sim2x2(title, Chisq_Pval, Gstat_Pval, Fisher_Pval, Fisher_midP;
        pmax = 0.051, tick = 0:0.005:1, figsize=(300, 300)
    )
    p = range(0, pmax, length=400)
    P2 = plot(title=title, size=figsize)
    plot!(titlefontsize=9, legendfontsize=8, tickfontsize=7, guidefontsize=8)
    plot!(xlabel="x", ylabel="probability of pval ≤ x")
    plot!(p, ecdf.(p; Y=Chisq_Pval);  label="χ²-test")
    plot!(p, ecdf.(p; Y=Gstat_Pval);  label="G-test",  ls=:dot, lw=1.5)
    plot!(p, ecdf.(p; Y=Fisher_Pval); label="Fisher",  ls=:dash)
    plot!(p, ecdf.(p; Y=Fisher_midP); label="mid-p F", ls=:dashdot)
    plot!([0, pmax], [0, pmax]; label="", color=:black, ls=:dot, alpha=0.5)
    plot!(xtick=tick, ytick=tick, xrotation=90)
    scatter!([0.01, 0.05], [0.01, 0.05]; label="", color=:red, markersize=3, markerstrokewidth=0)
end

function plot_sim2x2(rand_func, Λ; L=10^5, kwargs...)
    @time Chisq_Pval, Gstat_Pval, Fisher_Pval, Fisher_midP = sim2x2(rand_func, Λ; L=L)
    title = replace("$rand_func($Λ)", r"^rand_"=>"")
    sleep(0.1)
    plot_sim2x2(title, Chisq_Pval, Gstat_Pval, Fisher_Pval, Fisher_midP; kwargs...)
end

function plot_all2x2(Λ; L=10^5, figsize=(600, 600), kwargs...)
    P1 = plot_sim2x2(rand_Poissons, Λ; L=L, kwargs...)
    P2 = plot_sim2x2(rand_multinomial, Λ; L=L, kwargs...)
    P3 = plot_sim2x2(rand_2binomials, Λ; L=L, kwargs...)
    P4 = plot_sim2x2(rand_hypergeometric, Λ; L=L, kwargs...)
    plot(P1, P2, P3, P4; size=figsize, layout=grid(2,2))    
end
```

```julia
function simrxc(randrxc; L=10^5)
    Chisq_Pval  = Array{Float64, 1}(undef, L)
    Gstat_Pval  = Array{Float64, 1}(undef, L)
    for l in 1:L
        A = randrxc()
        Chisq_Pval[l]  = chisq_pval(A)
        Gstat_Pval[l]  = gstat_pval(A)
    end
    Chisq_Pval, Gstat_Pval
end

simrxc(rand_func, Λ; L=10^5) = simrxc(() -> rand_func(Λ); L=L)

function plot_simrxc(title, Chisq_Pval, Gstat_Pval;
        pmax = 0.051, tick = 0:0.005:1, figsize=(300, 300), kwargs...
    )
    p = range(0, pmax, length=400)
    P2 = plot(title=title, size=figsize)
    plot!(titlefontsize=8, legendfontsize=8, tickfontsize=7, guidefontsize=8)
    plot!(xlabel="x", ylabel="probability of pval ≤ x")
    plot!(p, ecdf.(p; Y=Chisq_Pval);  label="χ²-test")
    plot!(p, ecdf.(p; Y=Gstat_Pval);  label="G-test",  ls=:dot, lw=1.5)
    plot!([0, pmax], [0, pmax]; label="", color=:black, ls=:dot, alpha=0.5)
    plot!(xtick=tick, ytick=tick, xrotation=90)
    scatter!([0.01, 0.05], [0.01, 0.05]; label="", color=:red, markersize=3, markerstrokewidth=0)
end

function plot_simrxc(rand_func, Λ; L=10^5, kwargs...)
    @time Chisq_Pval, Gstat_Pval = simrxc(rand_func, Λ; L=L)
    title = replace("$rand_func($Λ)", r"^rand_"=>"")
    sleep(0.1)
    plot_simrxc(title, Chisq_Pval, Gstat_Pval; kwargs...)
end

function plot_allrxc(Λ; L=10^5, figsize=(600, 600), kwargs...)
    P1 = plot_simrxc(rand_Poissons, Λ; L=L, kwargs...)
    P2 = plot_simrxc(rand_multinomial, Λ; L=L, kwargs...)
    P3 = plot_simrxc(rand_prod_multinomials, Λ; L=L, kwargs...)
    P4 = Plots.Plot()
    plot(P1, P2, P3, P4; size=figsize, layout=grid(2,2))    
end
```
